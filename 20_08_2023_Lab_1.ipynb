```python
#singular value decomposition(SVD)
from skimage import data
import matplotlib.pyplot as plt
img=data.astronaut()
print(img.dtype,img.min(),img.max(),img.shape)
plt.imshow(img)
plt.show()
```

```python
#mormalizing it
img1=img/255.0
img1=np.transpose(img1,axes=(2,0,1))
```

```python
U,G,V=np.linalg.svd(img1)
print(U.shape,G.shape,V.shape)
```

```python
sigma=np.zeros((3,512,512))
for i in range(3):
  np.fill_diagonal(sigma[i,:,:],G[i,:])
print(sigma.shape)
```

```python
#reconstruction of image
reconstruct=U @ sigma @ V
reconstruct=np.transpose(reconstruct,axes=(1,2,0))
plt.imshow(reconstruct)
plt.show()
```

```python
#reconstruction by dimensionality reduction
k=26
reconst=U @ sigma[:,:,:k] @ V[:,:k,:]
reconst=np.transpose(reconst,axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```
